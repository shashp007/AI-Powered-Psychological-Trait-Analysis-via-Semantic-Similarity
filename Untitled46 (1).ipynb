```python
# Install required libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sentence_transformers import SentenceTransformer, losses
from torch.utils.data import DataLoader
import torch

# 1. DATA DEFINITION


# 18 NFC Questions (with reverse-scored indices)
nfc_questions = [
    "I would prefer complex to simple problems.",                              # Q1
    "I like to have the responsibility of handling a situation that requires a lot of thinking.",  # Q2
    "Thinking is not my idea of fun.",                                        # Q3 (Reverse)
    "I would rather do something that requires little thought than something that is sure to challenge my thinking abilities.",  # Q4 (Reverse)
    "I try to anticipate and avoid situations where there is likely a chance I will have to think in depth about something.",  # Q5 (Reverse)
    "I find satisfaction in deliberating hard and for long hours.",           # Q6
    "I only think as hard as I have to.",                                     # Q7 (Reverse)
    "I prefer to think about small, daily projects to long-term ones.",       # Q8 (Reverse)
    "I like tasks that require little thought once I've learned them.",       # Q9 (Reverse)
    "The idea of relying on thought to make my way to the top appeals to me.",  # Q10
    "I really enjoy a task that involves coming up with new solutions to problems.",  # Q11
    "Learning new ways to think doesn't excite me very much.",                # Q12 (Reverse)
    "I prefer my life to be filled with puzzles that I must solve.",          # Q13
    "The notion of thinking abstractly is appealing to me.",                  # Q14
    "I would prefer a task that is intellectual, difficult, and important to one that is somewhat important but does not require much thought.",  # Q15
    "I feel relief rather than satisfaction after completing a task that required a lot of mental effort.",  # Q16 (Reverse)
    "It's enough for me that something gets the job done; I don't care how or why it works.",  # Q17 (Reverse)
    "I usually end up deliberating about issues even when they do not affect me personally."   # Q18
]

# 10 Person X Statements
person_x_statements = [
    "I feel energized when solving puzzles that require deep concentration.",
    "Abstract thinking drains me—I prefer practical, straightforward tasks.",
    "I like seafood.",
    "I avoid discussions that involve theoretical concepts.",
    "Deliberating over multiple perspectives excites me.",
    "I prioritize efficiency over exploring new ways to solve problems.",
    "I enjoy breaking down ambiguous problems into smaller, manageable parts.",
    "I get frustrated when tasks require prolonged mental effort.",
    "I seek out intellectual challenges even in my free time.",
    "I prefer tasks with clear instructions rather than open-ended exploration."
]

# Reverse-scored question indices (0-based)
REVERSE_QS = [2, 3, 4, 6, 7, 8, 11, 15, 16]

# Keyword weights for score adjustment
KEYWORD_WEIGHTS = {
    "complex": 1.0, "thinking": 0.8, "deep": 0.7, "challeng": 0.6,  # Positive
    "simple": -1.0, "avoid": -1.0, "little": -0.8, "relief": -0.5    # Negative
}


# 2. SIMPLIFIED APPROACH (No Fine-Tuning)

# Initialize model (without fine-tuning to avoid complexity)
model = SentenceTransformer('all-MiniLM-L6-v2')


# 3. SEMANTIC SIMILARITY ANALYSIS


# Encode all texts
question_embeddings = model.encode(nfc_questions)
statement_embeddings = model.encode(person_x_statements)

# Compute similarity matrix
similarity = np.inner(question_embeddings, statement_embeddings)

# Adjust for reverse-scored questions
similarity[REVERSE_QS] = -similarity[REVERSE_QS]

# Scale from [-1, 1] to [0, 10]
scaled_scores = np.round((similarity + 1) * 5, 2)

# Apply keyword-based weighting
for i, question in enumerate(nfc_questions):
    for j, statement in enumerate(person_x_statements):
        for word, weight in KEYWORD_WEIGHTS.items():
            if word in question.lower() and word in statement.lower():
                scaled_scores[i][j] = np.clip(scaled_scores[i][j] + weight, 0, 10)

# 4. RESULTS & VISUALIZATION

# Create DataFrame
df = pd.DataFrame(
    scaled_scores,
    index=[f"Q{i+1}" for i in range(18)],
    columns=[f"S{j+1}" for j in range(10)]
)

# Print top 3 matches per question
print("TOP MATCHES (0-10 SCALE)")
for q in df.index:
    top3 = df.loc[q].nlargest(3)
    print(f"\n{q}: {nfc_questions[int(q[1:])-1]}")
    for s, score in top3.items():
        print(f"  {s}: {score:.1f} - {person_x_statements[int(s[1:])-1]}")

# Heatmap visualization
plt.figure(figsize=(12, 8))
sns.heatmap(
    df,
    annot=True,
    fmt=".1f",
    cmap="YlOrRd",
    cbar_kws={'label': 'Psychological Alignment Score (0-10)'},
    annot_kws={"size": 8}
)
plt.title("NFC Questions vs. Person X Statements\n(BERT Embeddings + Rules)", pad=20)
plt.xlabel("Person X Statements")
plt.ylabel("NFC Questions")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Save results
df.to_csv("nfc_statement_alignment.csv")
print("\nResults saved to 'nfc_statement_alignment.csv'")
```
